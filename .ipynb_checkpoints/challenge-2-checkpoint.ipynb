# 2. Get names and location of the 10 coffee places by passing Ironhack as your location

```python
def requests_for_foursquare (query, lat, lon, radius=500, limit=10):

    url = f"https://api.foursquare.com/v3/places/search?query={query}&ll={lat}%2C{lon}&radius={radius}&limit={limit}"

    headers = {
        "accept": "application/json",
        "Authorization": token
    }
    
    try:
        return requests.get(url, headers=headers).json()
    except:
        print("no :(")
```

```python
res_cof = requests_for_foursquare("coffee", lat, lon, limit=10)
res_cof
```

```python
def extracting_from_one_element(one):

    name = one["name"]
    address = one["location"]["address"]
    distance = one["distance"]

    
    small_dict = {
    "name": name,
    "address": address,
    "distance": distance,
}
    return small_dict
```

```python
def transform_request_into_df(res_cof):
    list_of_dictionaries = [extracting_from_one_element (element) for element in res_cof["results"]]
    df = pd.DataFrame(list_of_dictionaries)
    
    df.sort_values(by='distance', inplace = True)
    return df
```

```python
transform_request_into_df(res_cof)
```

# 4. Define a function that returns name, lat, long for starbucks OR restaurants AND for any given location

```python
def foursquare_any_given_location (venue, lat, lon):
    
    url = f"https://api.foursquare.com/v3/places/search?query={venue}&ll={lat}%2C{lon}&radius=500&limit=20"

    headers = {
        "accept": "application/json",
        "Authorization": token
    }
    
    
    if venue == "starbucks":
        star = requests.get(url, headers=headers).json()
        return star
    
    elif venue == "restaurants":
        rest = requests.get(url, headers=headers).json()
        return rest
    
    else:
        return "This is not the expected venue."
```
